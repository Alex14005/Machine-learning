```python
df_test = pd.read_csv('/content/test.csv')
df_train = pd.read_csv('/content/train.csv')
```

despues se revisan las caracteristicas de cada dataset, con describe()

```python
print(df_train.describe())
print('--------------------------')
print(df_test.describe())
```

aqui se hace el cambio de los datos de 'sexo' a números utilizando replace y tambien inplace=True, ese booleano quiere decir si quiere canbiar el df original o generar una copia

```python
df_train['Sex'].replace(['female','male'],[0,1],inplace=True)
df_test['Sex'].replace(['female','male'],[0,1],inplace=True)
```

```python
df_train['Embarked'].replace(['Q','S','C'],[0,1,2],inplace=True)
df_test['Embarked'].replace(['Q','S','C'],[0,1,2],inplace=True)
```

Despues se reemplaza los datos faltantes de la edad por la media de está columna

```python
print(df_train['Age'].mean())
print(df_test['Age'].mean())
```

se cambia los valores nulos por rpomedio

```python
promedio=30
df_train['Age'] = df_train['Age'].replace(np.nan, promedio)
df_test['Age'] = df_test['Age'].replace(np.nan, promedio)
```

Despues se crean varios grupos de acuerdo a bandasd e las edades como son:0-8,9-15,166-188,19-25,26-40,41-60,61-100

```python
bins = [0, 8, 15, 18, 25, 40, 60, 100]
names = ['1', '2', '3', '4', '5', '6', '7']
```

```python
df_train['Age'] = pd.cut(df_train['Age'], bins, labels=names)
```

en la siguiente opcion se elimina cabin, ya que tiene´muchos datos perdidos y sería muy complicado reemplazarlos

```python
df_train.drop(['Cabin'],axis=1,inplace=True)
df_test.drop(['Cabin'],axis=1,inplace=True)
```

después se descartan las columnas que concideramos que no van a influir en el analisis

```python
df_train = df_train.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
df_test = df_test.drop(['Name', 'Ticket'], axis=1)
```

Se elimina las filas con los datos perdiddo o nulos

```python
df_train.dropna(axis=0, how='any', inplace=True)
df_test.dropna(axis=0, how='any', inplace=True)
```

despues de la limpieza lo que se hace es verificar los datasets, para revisar que no nos falte lgún dato o revisar cuantos datoos  ns quedan como se ve abajo

```python
print(pd.isnull(df_train).sum())
print(pd.isnull(df_test).sum())
```

##Aplicacion de algoritmos de machine learning

estás son las librerias que usaremos para machine learning

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

depsues se separa la columna de sobrevivientes del dataset de entrenamiento

```python
X = np.array(df_train.drop(['Survived'], 1))
y = np.array(df_train['Survived'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

el siguiente algoritmo es de regresion logistica

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
Y_pred = logreg.predict(X_test)
print('Precisión Regresión Logística: ')
print(logreg.score(X_train,y_train))
```

Ahora se usa SVCSupport Vector Machine

```python
svc = SVC()
svc.fit(X_train, y_train)
Y_pred = svc.predict(X_test)
print('Precisión Soporte de Vectores: ')
print(svc.score(X_train, y_train))
```

algoritmo de vecinos mas cercanos K neighbors

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
Y_pred = knn.predict(X_test)
print('Precisión Vecinos más Cercanos')
print(knn.score(X_train, y_train))
```
