```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df['diagnosis'].replace(to_replace='M', value=1, inplace=True)
df['diagnosis'].replace(to_replace='B', value=0, inplace=True)
```

```python
df_data = df.copy()
```

```python
fig = plt.figure(figsize=(10,10))
df_data.corr()['diagnosis'].sort_values(ascending=True).plot(kind='bar')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df_data_scaled = scaler.fit_transform(df_data)
```

```python
df_data_scaled = pd.DataFrame(df_data_scaled)
```

```python
df_data_scaled.columns = df_data.columns
```

```python
df_data_scaled.head(5)
```

```python
sns.countplot(data=df_data, x='radius_mean', hue='diagnosis')
plt.show()
```

```python
def categoric_plot(column):
  fig = plt.figure(figsize=(10,10))
  sns.countplot(data=df_data, x=column, hue='diagnosis')
  plt.show()
```

```python
column_cat = df_data.select_dtypes(include='float64').columns
```

```python
for _ in column_cat:
  categoric_plot(_)
```

```python
fig = plt.figure(figsize=(10,10))
sns.pairplot(data=df_data, hue='diagnosis')
plt.show()
```

```python
X = df_data_scaled .drop('diagnosis', axis=1)
y = df_data_scaled['diagnosis'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
resutl = model.fit(x_train, y_train)
```

### continuación del proyecto

```python
from sklearn import metrics
prediction_test = model.predict(x_test)
print(metrics.accuracy_score(y_test, prediction_test))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
figur = plt.figure(figsize=(10,10))
cm = confusion_matrix(y_test, prediction_test, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=model.classes_)
disp.plot(cmap='gray')
```
